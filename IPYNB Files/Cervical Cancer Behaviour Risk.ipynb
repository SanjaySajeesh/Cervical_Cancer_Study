# <u>Task 3: Data Modelling</u>

###  <u>Task 3.1: Preliminary Steps to understand Classification Models</u>

####  <u>Task 3.1.1: Importing the Required Libraries</u>

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

####  <u>Task 3.1.2: Encoding Variables for Classification</u>

```python
from sklearn.preprocessing import LabelEncoder
for column in data.columns:
    if data[column].dtype == type(object):
        encoder = LabelEncoder()
        data[column] = encoder.fit_transform(data[column])
        
```

####  <u>Task 3.1.3: Assigning Variables for Classification</u>

```python
X = data.iloc[:,0:19].values
Y = data.ca_cervix
```

###  <u>Task 3.2: KNN Classification Modelling</u>

####  <u>Task 3.2.1: 50% Training and 50% Testing Split</u>

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.50, random_state=0)
k_Values = {}
k_Values_List= []
possible_K_Values = range(1,15)
for i in possible_K_Values:
    clf = KNeighborsClassifier(n_neighbors = i)
    clf.fit(X_train, y_train)
    predicted = clf.predict(X_test)
    k_Values[i] = metrics.accuracy_score(y_test, predicted)
    k_Values_List.append(k_Values[i])
    
plt.scatter(possible_K_Values, k_Values_List,marker='*')
plt.xlabel('K-Value')
plt.ylabel('Accuracy')
plt.show()
display(HTML('<b>Figure 3.2.1: Plotting the K-Values for 50/50 Split</b>'))
```

```python
clf = KNeighborsClassifier(8)
fit = clf.fit(X_train, y_train)
predicted = fit.predict(X_test)
Matrix_Confusion = confusion_matrix(y_test, predicted)
Error_Rate = (Matrix_Confusion[0][1] + Matrix_Confusion[1][0])/np.sum(Matrix_Confusion)

display(HTML('------------------------------<b><u>Confusion Matrix</b></u>-----------------------------------'))
print(Matrix_Confusion)
display(HTML('--------------------------------------------------------------------------------------'))
print ('\nThe Error Rate is: ', Error_Rate*100,'%','\n')
display(HTML('--------------------------------------------------------------------------------------'))
print (classification_report(y_test, predicted))
```
